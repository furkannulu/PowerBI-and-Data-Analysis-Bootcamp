```python
import yfinance as yf
import plotly.express as px
import plotly.graph_objects as go
import pandas as pd
import matplotlib.pyplot as plt
```

### 2021 EUR/TRY kur tablosunun incelenmesi

1. <i> Yahoo Finance</i> sitesinden 2021 yılına ait EUR/TRY kur değerlerini indiriniz

```python
Kur21 = yf.download('EURTRY=X', start = '2021-01-01', end = '2021-12-31')
B30_2021 = yf.download('XU030.IS', start = '2021-01-01', end = '2021-12-31')
```

2. Veriyi özet metodları ile inceleyiniz

```python
Kur21.head(10)
B30_2021.head(10)
```

```python
Kur21.describe()
B30_2021.describe()
```

```python
Kur21.info()
B30_2021.info()
```

3. Veri temizleme: <br>
	• Tabloda sadece "Adj Close" (düzeltilmiş kapanış) sütununu bırakınız.<br>
	• "Adj Close" sütununu " Kapanış (TRY)" olarak yeniden adlandırınız.

```python
Kur21 = Kur21[['Adj Close']]
# Kur21
Kur21.columns = ['Kapanış (TRY)']
Kur21
B30_2021 = B30_2021[['Adj Close']]
# B30_2021
B30_2021.columns = ['Kapanış (TRY)']
B30_2021
```

4. Tabloya yeni sütun ekleme:<br>
	• Tabloya yeni bir sütun ekleyerek Kapanış (TRY) değerlerini 1 Ocak 2021 tarihindeki kapanış değerine göre normalize ediniz.<br>
	• BIST30 endeksinin 2021 yılı boyunca değişimini yorumlayınız.


```python
Kur21['Normalize'] = Kur21['Kapanış (TRY)']/Kur21.loc[Kur21.index[0],'Kapanış (TRY)']
# Kur21
B30_2021['Normalize'] = B30_2021['Kapanış (TRY)']/B30_2021.loc[B30_2021.index[0],'Kapanış (TRY)']
# B30_2021
```

5. Veriyi grafik olarak inceleyiniz:<br>
	• Tabloyu px grafiği olarak görselleştiriniz.<br>
	• Tabloyu go. grafiği olarak görselleştiriniz.

```python
fig_1 = px.line(Kur21,y = 'Normalize',
        width = 1000, height = 500,
        title = 'Kur Endeksi 2021 - Normalize (1 Ocak 2021)',
        hover_data = {'Normalize' : ':.2f', 'Kapanış (TRY)' : ':.0f' },
        labels = {'Date' : 'Tarih' , 'Normalize' : 'Kur  Endeksi (EUR/TRY)' })
fig_1.show()
     
fig_2 = px.line(B30_2021,y = 'Normalize',
        width = 1000, height = 500,
        title = 'BIST30 Endeksi 2021 - Normalize (1 Ocak 2021)',
        hover_data = {'Normalize' : ':.2f', 'Kapanış (TRY)' : ':.0f' },
        labels = {'Date' : 'Tarih' , 'Normalize' : 'BIST30  Endeksi' })
fig_2.show()
     
```

# *Yukarıdaki grafikleri inceleyecek olursak;*

*   2021 Kur endeksi başlarda stabil durumda iken son çeyrekte 2 katlık bir sıçrama kaydetmiştir. Daha sonrasında cüzi bir miktar düşüş sağlamıştır.
*   BIST30 endeksi 2021 yılının ilk çeyreği stabil iken daha sonrasında bir miktar düşme kaydetmiş ama buna rağmen kendisini eski düzeyine getirmiştir. Son çeyrekte ise Kurendeksinde olduğu gibi 2 katlık bir anlık sıçarama kaydetmiş ve sonrasında belirli bir değer aralığında sabit kalmıştır.


```python
fig = go.Figure()
fig.add_trace(go.Bar( x=B30_2021.index, y=B30_2021['Normalize'], name = 'Normalize BIST30'))

fig.update_layout(title='Normalize BIST30',
                  xaxis_title = 'Tarih', xaxis_dtick = 'M1',
                  yaxis_title = 'Kapanış (TRY)')

fig1 = go.Figure()
fig1.add_trace(go.Bar( x=B30_2021.index, y=B30_2021['Normalize'], name = 'Normalize BIST30'))

fig1.update_layout(title='Normalize BIST30',
                  xaxis_title = 'Tarih', xaxis_dtick = 'M1',
                  yaxis_title = 'Kapanış (TRY)')
#Daha önceki yönergelerde merge etmemiz istenmediği için  go plotlarını ayrı ayrı çizdirdim.
```

### 2021 ve 2022 değerlerini birlikte inceleme

1. <i> Yahoo Finance</i> sitesinden 2022 BIST30 ve EUR/TRY kur değerlerini indiriniz

```python
B30_2022 = yf.download('XU030.IS', start = '2022-01-01', end = '2022-12-31')
Kur22 = yf.download('EURTRY=X', start = '2022-01-01', end = '2022-12-31')
```

2. Indirdiğiniz verileri 2021 verilerini temizleyip düzenlediğiniz üzere hazırlayınız.

```python
Kur22 = Kur22[['Adj Close']]
Kur22
B30_2022 = B30_2022[['Adj Close']]
B30_2022
```

```python
Kur22.columns = ['Kapanış (TRY)']
# Kur
Kur22['Normalize'] = Kur22['Kapanış (TRY)']/Kur22.loc[Kur22.index[0],'Kapanış (TRY)']
# Kur

B30_2022.columns = ['Kapanış (TRY)']
# B30_2022
B30_2022['Normalize'] = B30_2022['Kapanış (TRY)']/B30_2022.loc[B30_2022.index[0],'Kapanış (TRY)']
# B30_2022
```

3. 2021 ve 2022 veri tablolarını tek bir tablo haline getiriniz.

```python
Veri_B30 = pd.concat([B30_2021, B30_2022], axis=0)
Veri_B30
Veri_Kur = pd.concat([Kur21, Kur22], axis=0)
Veri_Kur
```

4. Veriyi grafik olarak inceleyiniz:<br>
	• Tabloları px grafiği olarak görselleştiriniz.<br>
	• Tabloları go. grafiği olarak görselleştiriniz.<br>
	• Tabloları subplots grafiği olarak görselleştiriniz.

```python
fig_1 = px.line(Veri_B30,y = 'Normalize',
        width = 1000, height = 500,
        title = 'BIST30 Endeksi 2021/2022',
        hover_data = {'Normalize' : ':.2f', 'Kapanış (TRY)' : ':.0f' },
        labels = {'Date' : 'Tarih' , 'Normalize' : 'BIST30  Endeksi' })
fig_1.show()
     
fig_2 = px.line(Veri_Kur,y = 'Normalize',
        width = 1000, height = 500,
        title = 'Kur Endeksi 2021/2022',
        hover_data = {'Normalize' : ':.2f', 'Kapanış (TRY)' : ':.0f' },
        labels = {'Date' : 'Tarih' , 'Normalize' : 'Kur  Endeksi (EUR/TRY)' })
fig_2.show()
```

# *Yukarıdaki grafikleri inceleyecek olursak ;*

*   BIST30 endeksi 2021 yılında çok değişime uğramamış fakat 2022 2. yarısında sürekli bir artış durumuna geçmiştir
*   Kur endeksi 2021 sonuna kadar fazla bir yükseliş kaydetmezken, 2021 sonuna gelindiğinde büyük bir sıçrama yapmıştır. Fakat 2022 başlarında aynı seviyelere hatta daha da aşağılara gerilemiştir. İlerleyen zamanlarda yükseliş kaydetse de bu yükseliş miktar olarak fazla değil ama sürekli olmuştur.



```python
fig = go.Figure()
fig.add_trace(go.Bar( x=Veri_B30.index, y=Veri_B30['Normalize'], name = '2021/2022 Normalize BIST30'))

fig.update_layout(title='2021/2022 Normalize BIST30',
                  xaxis_title = 'Tarih', xaxis_dtick = 'M1',
                  yaxis_title = 'Kapanış (TRY)')

fig1 = go.Figure()
fig1.add_trace(go.Bar( x=Veri_Kur.index, y=Veri_Kur['Normalize'], name = '2021/2022 Normalize Kur EUR/TRY'))

fig1.update_layout(title='2021/2022 Normalize Kur EUR/TRY',
                  xaxis_title = 'Tarih', xaxis_dtick = 'M1',
                  yaxis_title = 'Kapanış (TRY)')
```
